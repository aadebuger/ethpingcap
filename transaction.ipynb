```python
import pandas as pd
filename='/home/aadebuger/GEXTMAC/ethereum1t/etldata/transactions.csv'
tv = pd.read_csv(filename)
```

```python
from sqlalchemy import create_engine
engine = create_engine("mysql+pymysql://{}:{}@{}/{}".format('root', '', 'localhost:4000', 'testexample'))
con = engine.connect()

print("test1")
```

```python
print(len(tv))
l=0
r=500
length =len(tv)
while(l<length):
    tv[l:r].to_sql(name='transcation', con=con, if_exists='append', index=False,chunksize=1000)
    print("to_sql ok",l)
    l+=500
    r+=500
```
